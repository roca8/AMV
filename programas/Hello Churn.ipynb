```python
var = ['trip_start_timestamp',
       'tips',
       'trip_seconds',
       'trip_miles',
       'trip_total',
       'payment_type','taxi_id']
```

```python
%%time
l = []
for a in lst_arch:
    print(a)
    df = pd.read_csv(os.path.join('data',a))[var]
    df = df.merge(aux,on='taxi_id',how='inner')
    l.append(df)
```

```python
df = pd.concat(l,ignore_index=False)
```

```python
df.head()
```

```python
df['w']=pd.to_datetime(df['trip_start_timestamp']).dt.date.map(lambda x:x.strftime('%V')).astype(int)
```

```python
aux = df.drop(['payment_type','trip_start_timestamp'],axis=1).groupby(['taxi_id','w']).agg(['count','sum'])
```

```python
aux.reset_index(inplace=True)
```

```python
aux.head()
```

```python
aux.columns = [a+'_'+b for a,b in aux.columns]
```

```python
var = ['tips_count','tips_sum','trip_seconds_sum', 'trip_miles_sum', 'trip_total_sum']
```

```python
aux = aux[var+['taxi_id_','w_']]
```

```python
aux
```

```python
t = aux[['taxi_id_','w_']].copy()
for k in range(4):
    aux['w0'] = aux['w_']+k
    t = t.merge(aux.drop('w_',axis=1),left_on=['taxi_id_','w_'],right_on=['taxi_id_','w0'],how='left')
    t.rename(columns=dict(zip(var,['%s_%d'%(v,k+1) for v in var])),inplace=True)
    t.drop('w0',axis=1,inplace=True)
```

```python
t.head()
```

```python
for v in var:
    for k in range(4):
        l_feat = ['%s_%d'%(v,i+1) for i in range(k+1)]
        print(l_feat,v)
        t['v_total_%s_%d'%(v,k+1)] = t[l_feat].sum(axis=1)
    t.drop(l_feat,axis=1,inplace=True)
```
