## 4. CreaciÃ³n de variable objetivo para clasificaciÃ³n

```python
df['target'] = (df.d_01=='Actuaria').astype(int)
```

## 5. PreparaciÃ³n conjunto de entrenamiento

```python
X = df[[v for v in df.columns if v[:2]=='c_']].copy()
```

```python
(1-X.describe().T['count']/len(X)).sort_values()
```

```python
X['c_06'].fillna(0,inplace=True)
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
Xm = pd.DataFrame(sc.fit_transform(X),columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df[['target']].copy()
```

```python
Xt, Xv, yt, yv = train_test_split(Xm,y,train_size=0.7)
```

## 6. HiperparametrizaciÃ³n

```python
from sklearn.svm import SVC
```

```python
modelo = SVC(kernel='linear')
```

```python
param = dict(C=np.arange(0,3,0.01))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=modelo,verbose=False,
                    param_grid=param,
                    n_jobs=-1,
                    cv=3,
                    scoring='accuracy')
```

```python
%%time
grid.fit(Xm,y)
```

```python
modelo = grid.best_estimator_
```

```python
grid.best_score_,grid.best_params_
```

## 7. Entrenamiento Final

```python
modelo.fit(Xt,yt)
```

```python
from sklearn.metrics import accuracy_score
```

```python
train = accuracy_score(y_pred=modelo.predict(Xt),y_true=yt)
val = accuracy_score(y_pred=modelo.predict(Xv),y_true=yv)
print(train)
print(val)
```

```python
%%time
k = 0
while abs(train-val)>0.01:
    k+=1
    Xt, Xv, yt, yv = train_test_split(Xm,y,train_size=0.7)
    grid.fit(Xm,y)
    modelo = grid.best_estimator_
    modelo.fit(Xt,yt)
    train = accuracy_score(y_pred=modelo.predict(Xt),y_true=yt)
    val = accuracy_score(y_pred=modelo.predict(Xv),y_true=yv)
    print(train)
    print(val)
```
